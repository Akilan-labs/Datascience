```python
import numpy as np
import pandas as pd

np.random.seed(87)

data = {
    "Players":np.arange(1,12),
    "Runs":np.random.randint(0,150,size=11)
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df["Players"], df["Runs"],marker='o',color='Green')
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Players vs Runs Plot")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Players"], df["Runs"], marker='*', color='green',linestyle='--')
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Line Plot: Players vs Runs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Players"], df["Runs"], marker='*', color='green',linestyle=':')
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Line Plot: Players vs Runs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Players"], df["Runs"], marker='o', color='orange',linestyle='--')
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Line Plot: Players vs Runs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Players"], df["Runs"], color='pink')
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Vertical Bar Plot: Players vs Runs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.barh(df["Players"], df["Runs"], color='green')
plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Vertical Bar Plot: Players vs Runs")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(df["Runs"].fillna(0), labels=df["Players"], autopct='%1.1f%%', startangle=90,
        colors=plt.cm.Pastel1(np.arange(len(df))))
plt.title("Pie Chart: Runs Distribution Among Players")
plt.grid(True)
plt.show()
```
