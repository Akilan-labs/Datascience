```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0, 11)
y = x

plt.plot(x, y, marker='o')
plt.title("Basic Line Graph")
plt.xlabel("x")
plt.ylabel("y = x")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ["A", "B", "C", "D"]
sales = [40, 70, 30, 90]

plt.bar(products, sales)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)

plt.figure(figsize=(10, 8))

# Line plot
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Line Plot")

# Scatter plot
plt.subplot(2, 2, 2)
plt.scatter(x, np.cos(x))
plt.title("Scatter Plot")

# Bar plot
plt.subplot(2, 2, 3)
plt.bar(["A", "B", "C"], [3, 5, 2])
plt.title("Bar Chart")

# Pie chart
plt.subplot(2, 2, 4)
plt.pie([20, 30, 50], labels=["Red", "Blue", "Green"], autopct="%1.1f%%")
plt.title("Pie Chart")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, 200)

plt.hist(data, bins=20)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
